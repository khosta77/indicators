```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gazp = pd.read_csv("./GAZP.csv",delimiter=';')
gazp.columns = gazp.columns.str.replace('<', '')
gazp.columns = gazp.columns.str.replace('>', '')
gazp.head()
```

```python
gazp['SMA30'] = gazp["OPEN"].rolling(window=30).mean ()
gazp.to_csv("GAZP_PY.csv", sep=';')
gazp.tail()
```

```python
MySMA = pd.read_csv("./gazp_sma30.csv",delimiter=';')
MySMA.tail()
```

```python
plt.plot(gazp["SMA30"])
plt.plot(gazp["OPEN"])
plt.plot(MySMA["SMA30"], "--")
plt.show()
```
